```python
reference_vocab=None
for i in range(len(clusters)):
    if len(clusters[i]) <= 2:
        continue
    for atom in clusters[i]:
        for j in nei_list[atom]:
            if i >= j or len(clusters[j]) <= 2:
                continue
            inter = clusters[i] & clusters[j]
            if len(inter) >= 2:
                merge = clusters[i] | clusters[j]
                if reference_vocab is not None:
                    smile_merge = Chem.MolFragmentToSmiles(mol, merge, canonical=True, kekuleSmiles=True)
                    if reference_vocab[smile_merge] <= 99:
                        continue
                clusters[i] = merge
                clusters[j] = set()
```

```python
# 读取元素周期表
ptable = Chem.GetPeriodicTable()
# 原子序数，sdf文件中的三维坐标
element, pos = [], []
# 初始化加总位置accumulated_position为三维全0，加总位置是坐标乘上相对原子质量
accum_pos = np.array([0.0, 0.0, 0.0], dtype=np.float32)
# 初始化加总相对原子质量为0.0
accum_mass = 0.0
# 对于sdf中的每一行，坐标信息从第四行开始，每一行都从下标4开始；有多少个原子，坐标信息就有多少行
for atom_line in map(lambda x:x.split(), sdf[4:4+num_atoms]):
    # 按照空格分每一行x.split，前三个（0-2）是坐标信息，字符串转为浮点数
    x, y, z = map(float, atom_line[:3])
    # 第3个字段是元素符号
    symb = atom_line[3]
    # 元素符号大写，再获取在元素周期表中的原子序数
    atomic_number = ptable.GetAtomicNumber(symb.capitalize())
    # 添加的是每个原子在元素周期表中的原子序数
    element.append(atomic_number)
    # 位置列表记录sdf中提取出来的三维坐标
    pos.append([x, y, z])

    # 获取原子在元素周期表中的相对原子质量
    atomic_weight = ptable.GetAtomicWeight(atomic_number)
    # 加总位置是增加原子的坐标乘上相对原子质量
    accum_pos += np.array([x, y, z]) * atomic_weight
    # 加总质量是增加原子的相对原子质量
    accum_mass += atomic_weight
```

```python

def reset_moltree_root(moltree, ligand_pos, protein_pos):
    # (11, 3)，小分子含11个原子，11个原子的三维坐标
    # (631, 3)，蛋白质含631个原子，631个原子的三维坐标

    # 计算配体和蛋白质原子两两坐标距离（没有开方，因为只用来比大小）
    # 配体（小分子）所有原子三维坐标，都先取平方再相加
    ligand2 = np.sum(np.square(ligand_pos), 1, keepdims=True)  # [11, 1]
    # 蛋白质口袋所有原子三维坐标，都先取平方再相加
    protein2 = np.sum(np.square(protein_pos), 1, keepdims=True)
    # 坐标之差取平方求和结果
    dist = np.add(np.add(-2 * np.dot(ligand_pos, protein_pos.T), ligand2), protein2.T)  # 得到(11, 631)

    # 每个配体小分子原子 和最近的蛋白质原子的距离平方
    min_dist = np.min(dist, 1)  # [11,]
    avg_min_dist = []  # 配体小分子树结点的最小距离平方
    # 配体小分子构建的最小生成树，结点是键和长度不超过8的环（cluster），边是cluster之间交集原子个数但是无边权，最小生成树尽可能保留了交集原子较多的那些边
    # clique是 键或者长度不超过8的环（cluster）的原子索引集合
    for node in moltree.nodes:
        # 先取到clique里面的每个原子，每个原子都有距离平方，clique里面的最小距离平方作为这个cluster键或者环的最小距离平方
        avg_min_dist.append(np.min(min_dist[node.clique]))

    # 在mol_tree里面建树的时候，root考虑的是包含0号原子的cluster，而且要把root弄到最前面，0号结点
    # 现在把root变成和蛋白质口袋有最小距离的cluster
    root = np.argmin(avg_min_dist)
    if root > 0:  # 要把root结点提到第0个
        moltree.nodes[0], moltree.nodes[root] = moltree.nodes[root], moltree.nodes[0]
    
    # 拥有和蛋白质口袋最小距离的键或者环（根结点），里面的原子，和蛋白质口袋原子两两有距离
    # 对每个蛋白质口袋原子，取和根结点cluster里面原子最小的两两距离，再在所有最小两两距离里面取一个最小的，得到的索引是蛋白质口袋原子的索引
    contact_idx = np.argmin(np.min(dist[moltree.nodes[0].clique], 0))  # 键是两个原子：[2, 631]->[631,]->[]标量
    # 每个蛋白质原子，如果有配体小分子的某个原子距离它小于4，就认为是contact_protein，就是接触蛋白质原子
    contact_protein = torch.tensor(np.min(dist, 0) < 4 ** 2)  # [631]


    # 返回，更新了根结点的配体小分子生成树（从原子来看根结点和蛋白质口袋最近）
    # 蛋白质原子是否为接触原子（和配体小分子某个原子距离小于4）
    # 接触蛋白质索引（和根结点最近的蛋白质原子索引）
    return moltree, contact_protein, torch.tensor([contact_idx])
```

```python
# 接下来是不是很熟悉的bfs遍历
bfs_queue = [0]
bfs_perm = []
bfs_focal = []
visited = {bfs_queue[0]}
while len(bfs_queue) > 0:  # 队列不空
    current = bfs_queue.pop(0)  # 出队
    bfs_perm.append(current)  # 遍历顺序
    next_candid = []  # 下一个层次还要随机打乱
    # 跑链（所有邻居）
    for motif in moltree.nodes[current].neighbors:  # 当前结点的邻居
        print("neighbor", motif.nid)
        if motif.nid in visited: continue  
        next_candid.append(motif.nid)  # 没有访问过，记录，准备入队列
        visited.add(motif.nid)  # 访问
        bfs_focal.append(current)  # 如果它的邻居都访问过了，就不会到这一步，就不会成为加到bfs_focal，这是什么情况
        # 在bfs中，记录新访问的结点是由谁引出来的，重复的不算
    print("visited", visited)
    print("next_candid", next_candid)
    print("bfs_focal", bfs_focal)
    random.shuffle(next_candid)  # 下一个层次先随机打乱，再入队列
    bfs_queue += next_candid
```

```python
def get_submol(mol, idxs, mark=[]):
    # 在enumerate_assemble中，ctr_mol = get_submol(mol, idxs, mark=current.clique) 调用
    # mol: data['moltree'].mol 分子moltree.mol对象
    # idx: # contact_idx 没有被mask的小分子原子索引列表 
    # current:  moltree.nodes[bfs_focal[-num_masked]] 现在能看到的就是bfs_focal的排除后面那num_masked个结点，剩下的那些

    # 从名字来看，好像是获得子图。
    # 就是利用这些记录好的原子索引列表，回到原来完整整全的那个mol里面，去取出对应的原子和键
    # 当然，取出来是复制出新的，也就是构造出新的

    new_mol = Chem.RWMol(Chem.MolFromSmiles(''))  # Chem.RWMol用于分子读写的类，可编辑和修改分子，原来还可以从空字符串创建分子
    map = {}  # 什么映射呢。在整体mol里面的那个原子索引编号，在new_mol里面的原子索引
    for atom in mol.GetAtoms():  # 对于小分子整体mol的每个原子
        if atom.GetIdx() in idxs:  # 如果原子索引在contact_idx里面，contact_idx是没有被mask的所有可见片段的小分子原子索引
            new_atom = copy_atom(atom)  # 复制原子
            if atom.GetIdx() in mark:  # 如果在current.clique里面，也就是当前current锁定到的那一个可见片段的小分子原子索引
                new_atom.SetAtomMapNum(1)  # 标记原子序号为1
            else:  # 只是可见，但不是current focal
                new_atom.SetAtomMapNum(0)  # 标记为0，原子的序号是从0开始的，在图中不会标出0
            # 原子索引在可见片段里面，还要加进map，把复制的新原子加进map，复制的新原子携带了一些1，0信息
            # 在整体mol里面的那个原子索引编号，在new_mol里面的原子索引
            map[atom.GetIdx()] = new_mol.AddAtom(new_atom)  # 要添加的原子对象 返回值是新添加原子的索引
    
    for bond in mol.GetBonds():  # 对于小分子整体mol的每个键
        a1 = bond.GetBeginAtom().GetIdx()  # 键的起始原子索引
        a2 = bond.GetEndAtom().GetIdx()  # 键的终止原子索引
        if a1 in idxs and a2 in idxs:  # 如果起始原子和终止原子都在可见片段里面
            bt = bond.GetBondType()  # 获取键的类型
            new_mol.AddBond(map[a1], map[a2], bt)  # new_mol就相应加上这个键

    # 返回的是contact_idx里面的，也就是所有可见片段的整全情况，有标记为1的，就是current的focal motif（正如形式参数名mark的意义）
    return new_mol.GetMol()  # 返回对象
```

```python
def atom_equal(a1, a2):
    # 从元素符号和电荷两方面判断两个原子是否相等
    return a1.GetSymbol() == a2.GetSymbol() and a1.GetFormalCharge() == a2.GetFormalCharge()


# Bond type not considered because all aromatic (so SINGLE matches DOUBLE)
def ring_bond_equal(bond1, bond2, reverse=False):
    # 比较两个化学键是否相同：起始原子和键类型
    # 第一个键的起始原子元组
    b1 = (bond1.GetBeginAtom(), bond1.GetEndAtom())
    if reverse:  # 如果反过来，起变终，终变起
        b2 = (bond2.GetEndAtom(), bond2.GetBeginAtom())
    else:
        b2 = (bond2.GetBeginAtom(), bond2.GetEndAtom())
    # 如果起始原子分别一样，并且键类型也一样，就认为是一样
    return atom_equal(b1[0], b2[0]) and atom_equal(b1[1], b2[1]) and bond1.GetBondType() == bond2.GetBondType()
```

```python
def attach(ctr_mol, nei_mol, amap):
    # cand_mol, _ = attach(ctr_mol, next.mol, amap) 
    # ctr_mol: rdkit mol对象，现在可见的分子情况，是当前focal的原子特别标记为1
    # nei_mol: next.mol moltree.nodes[bfs_perm[-num_masked]] -num_masked往后是masked个看不到的，按照bfs顺序bfs_perm比较靠后的那几个结点（团簇）    
    # amap: 提取自enum_attach的，有希望合二为一的原子和键字典列表
    ctr_mol = Chem.RWMol(ctr_mol)
    for atom in nei_mol.GetAtoms():
        if atom.GetIdx() not in amap:
            new_atom = copy_atom(atom)
            new_atom.SetAtomMapNum(2)
            amap[atom.GetIdx()] = ctr_mol.AddAtom(new_atom)
    display(ctr_mol)
#     print(amap)
    for bond in nei_mol.GetBonds():
        a1 = amap[bond.GetBeginAtom().GetIdx()]
        a2 = amap[bond.GetEndAtom().GetIdx()]
        if ctr_mol.GetBondBetweenAtoms(a1, a2) is None:
            ctr_mol.AddBond(a1, a2, bond.GetBondType())
        display(ctr_mol)
    return ctr_mol.GetMol(), amap
```

```python

class LigandCountNeighbors(object):

    @staticmethod
    def count_neighbors(edge_index, symmetry, valence=None, num_nodes=None):
        assert symmetry == True, 'Only support symmetrical edges.'  # 无向的，边是对称，之前在建树的时候专门有操作，一个变两个，而且还有一个巧妙的排序方式

        if num_nodes is None:  # 如果没有配体小分子原子个数，这种情况不知道是什么时候出现
            num_nodes = maybe_num_nodes(edge_index)

        if valence is None:  # 如果没有提供键的类型
            valence = torch.ones([edge_index.size(1)], device=edge_index.device)
        valence = valence.view(edge_index.size(1))  # 规整个数，和sdf解析出来的键个数两倍要一致，如果valence是None，本身就是这么创建的
        # torch.tensor.view返回了新数组 Returns a new tensor with the same data as the self tensor but of a different shape.
        # scatter_add是在做什么呢，在统计每个结点的度（边对称无向），再转化为long类型（int64）
        return scatter_add(valence, index=edge_index[0], dim=0, dim_size=num_nodes).long() 

    def __init__(self):
        super().__init__()

    def __call__(self, data):
        data['ligand_num_neighbors'] = self.count_neighbors(  # 小分子每个原子的度（边对称无向），名称取得很好，配体邻居数量
            data['ligand_bond_index'],  # 边索引，就是原子索引对，sdf解析是一变二了的，而且排好了序
            symmetry=True,  # 对称
            # 没有传valence
            num_nodes=data['ligand_element'].size(0),  #  配体小分子的原子个数，配体小分子原子在元素周期表中的原子序数ndarray
        )
        data['ligand_atom_valence'] = self.count_neighbors(  # 考虑了键类型的结点度，相当于度里面有键类型的“权重”
            data['ligand_bond_index'],
            symmetry=True,  # 对称
            valence=data['ligand_bond_type'],  # 边类型，就是键类型的索引 (RDKit的键类型有22种，这里只用到其中的四种)
            num_nodes=data['ligand_element'].size(0),  #  配体小分子的原子个数，配体小分子原子在元素周期表中的原子序数ndarray
        )
        return data
```

```python
def batch_dihedrals(p0, p1, p2, p3, angle=False):
# data['true_sin'], data['true_cos'] = batch_dihedrals(
#    xn_pos[xn_idx], x_pos.repeat(9, 1), y_pos.repeat(9, 1), yn_pos[yn_idx])
    s1 = p1 - p0
    s2 = p2 - p1
    s3 = p3 - p2

    sin_d_ = torch.linalg.norm(s2, dim=-1) * torch.sum(s1 * torch.cross(s2, s3, dim=-1), dim=-1)
    cos_d_ = torch.sum(torch.cross(s1, s2, dim=-1) * torch.cross(s2, s3, dim=-1), dim=-1)

    if angle:
        return torch.atan2(sin_d_, cos_d_ + 1e-10)

    else:
        den = torch.linalg.norm(torch.cross(s1, s2, dim=-1), dim=-1) * torch.linalg.norm(torch.cross(s2, s3, dim=-1), dim=-1) + 1e-10
        return sin_d_/den, cos_d_/den
```

```python
src, index
#index按行遍历（tensor内部）
#index的行列下标决定src的行列下标，index的值决定out里面第二维的索引(dim=-1)，index的行下标决定out里面第一维的索引
# out[0=行下标，0=value（0, 3=列下标）] = src[0,0]+src[0,3]=1+4，0是index的索引（第一行内部dim=0），0是index的值（dim=-1），0和3是对应值为0的索引，去src里面取值1和4，相加
# out[0=行下标，1=value（1=列下标）] = src[0,1]=2，0是index的索引（第一行内部dim=0），1是index的值（dim=-1），1是对应为值1的索引，去src里面取值2
# out[0=行下标，2=value（2=列下标）] = src[0,2]=3，0是index的索引（第一行内部dim=0），2是index的值（dim=-1），2是对应为值2的索引，去src里面取值3
# out[0=行下标，3=value（4=列下标）] = src[0,4]=5，0是index的索引（第一行内部dim=0），3是index的值（dim=-1），4是对应为值3的索引，去src里面取值5

# out[1=行下标，0=value（0=列下标）] = src[1,0]=6，1是index的索引（第二行内部dim=0），0是index的值（dim=-1），0是对应为值0的索引，去src里面[1,0]取值6
# out[1=行下标，1=value（1,2=列下标）] = src[1,1]+src[1,2]=7+8
# out[1=行下标，2=value（3,4=列下标）]= src[1,3]+src[1,4]=9+10
# out[1,3]? = 0  # fill_value=0
```
